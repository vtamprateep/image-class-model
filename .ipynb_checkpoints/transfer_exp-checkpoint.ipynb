```python
# Import modules
from tensorflow import keras
from sklearn import preprocessing
from sklearn import model_selection
import pandas
import numpy
```

```python
# Import data
dataset = pandas.read_csv('./data/data_labels.csv', names = ['image_id', 'label'])
dataset = dataset.to_numpy()
numpy.random.shuffle(dataset)
```

```python
# Split numpy array
split_data = numpy.hsplit(dataset, 2)
image_id = numpy.ravel(split_data[0])
y = numpy.ravel(split_data[1])

# Encode labels
le = preprocessing.LabelEncoder()
le.fit(y)
y = le.transform(y)
image_classes = le.classes_
```

```python
# Load all images in numpy array
X = numpy.empty((len(image_id), 160, 160, 3))

for enum, i in enumerate(image_id):
    image_path = './data/images/' + i
    image_arr = keras.preprocessing.image.load_img(image_path, color_mode = 'rgb', target_size = (160, 160, 3))
    image_arr = keras.preprocessing.image.img_to_array(image_arr)
    X[enum,] = image_arr
    
X = X / 255 # Normalize
```

```python
# Base mode is MobileNetV2

base_model = keras.applications.MobileNetV2(input_shape = (160, 160, 3), include_top = False, weights = 'imagenet')
base_model.trainable = False
base_model.summary()
```

```python
model = keras.Sequential()
model.add(base_model)
model.add(keras.layers.GlobalAveragePooling2D())
model.add(keras.layers.Dense(6, activation = 'softmax'))
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'Adam', metrics = ['accuracy'])
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.2)
```

```python
model.fit(X_train, y_train, epochs = 5, validation_data = (X_test, y_test))
```
